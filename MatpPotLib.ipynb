### Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter

%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data

x =[1,2,3,4]
y= [11,22,33,44]

fig, ax = plt.subplots(figsize=(10,10))

# 3. Plot data

ax.plot(x,y)

# 4. customize plot

ax.set(title="Simple plot", xlabel="x-axis", ylabel="y-axis")

# 5. save & show (you save the whole figure)

fig.savefig('images/sample-plot.png')
```

```python
# make a plot form a dictionary

nut_butter_prices = {"Almond butter": 10, "Peanut butter": 8,
                     "Cashew butter": 12}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),height= nut_butter_prices.values())
ax.set(title="Mwibutsa's Nut Butter Store",
       ylabel="Price ($)");
```

# Two options for subplots

```python
# subplot option 1

fig, ((ax1, ax2), (ax3,ax4)) = plt.subplots(nrows=2,
                                           ncols=2,
                                           figsize=(10,5))


# Plot to each different axis

ax1.plot(x,(x))
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000))
```

```python
# Subplotfig,  option 2

fig,ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

# Plot to each different index

ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1,1].hist(np.random.randn(1000))

```

### Which one should we use? (pyplot vs matplotlib OO method?

* When plotting something quickly, It's okay to use pyplot method
* when plotting something more advanced, use matplotlib OO Method


```python
 heart_disease
```

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50
```

```python
len(over_50)
```

```python
over_50.head()
```

```python
# Pyplot method
over_50.plot(kind='scatter', x='age', y='chol', c='target');
```

```python
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
             x='age', y='chol',
             c='target', ax=ax);

ax.set_xlim([45,100])
```

```python
## OO Method from scratch

fig, ax = plt.subplots(figsize=(10,6))
# Plot the data
scatter = ax.scatter(x=over_50['age'], y=over_50['chol'],
                    c=over_50['target'])

ax.set(title="Heart Disease and Cholersterol levels",
      xlabel="Age", ylabel="Cholersterol")

# Add a legend

ax.legend(*scatter.legend_elements(), title="Target")

ax.axhline(over_50['chol'].mean(), linestyle="dashed")
```

```python
over_50.head()
```

```python
# Subplot of a chol, age. thalach

fig,(ax0, ax1) = plt.subplots(figsize=(10,10) , 
                              nrows=2, 
                              ncols=1,
                             sharex=True)


# add Data to ax0
scatter0 = ax0.scatter(x=over_50['age'],
                    y=over_50['chol'], 
                    c=over_50['target'])


# Customize ax0

ax0.set(title="Heart Disease and cholesterol Levels",
       ylabel="Cholesterol")

ax0.legend(*scatter0.legend_elements(),title="Target")

# Add a mean line

ax0.axhline(y=over_50['chol'].mean(), linestyle='--')



# Add Data to Ax1

scatter1 = ax1.scatter(x=over_50['age'], 
                       y=over_50['thalach'],
                      c=over_50['target'])

# Customize ax1

ax1.set(title="Heart Disease and Max heart rate",
       xlabel="Age",
        ylabel="Max heart rate")

ax1.legend(*scatter1.legend_elements(), title="Target")

# Add mean line

ax1.axhline(y=over_50['thalach'].mean(), linestyle='--');



# Add a title to the figure

fig.suptitle("Heart Disease Analysis", fontsize=20,
             fontweight='bold');
```

```python
# set the style

plt.style.use('seaborn-whitegrid')

# Subplot of a chol, age. thalach

fig,(ax0, ax1) = plt.subplots(figsize=(10,10) , 
                              nrows=2, 
                              ncols=1,
                             sharex=True)


# add Data to ax0
scatter0 = ax0.scatter(x=over_50['age'],
                    y=over_50['chol'], 
                    c=over_50['target'],
                      cmap='winter')


# Customize ax0

ax0.set(title="Heart Disease and cholesterol Levels",
       ylabel="Cholesterol")

ax0.set_xlim(50,80)

ax0.legend(*scatter0.legend_elements(),title="Target")

# Add a mean line

ax0.axhline(y=over_50['chol'].mean(), linestyle='--')



# Add Data to Ax1

scatter1 = ax1.scatter(x=over_50['age'], 
                       y=over_50['thalach'],
                      c=over_50['target'],
                      cmap='winter')

# Customize ax1

ax1.set(title="Heart Disease and Max heart rate",
       xlabel="Age",
        ylabel="Max heart rate")

# change ax1 x axis limits

ax1.set_xlim(50,80)
ax1.set_ylim(60,200)
ax1.legend(*scatter1.legend_elements(), title="Target")

# Add mean line

ax1.axhline(y=over_50['thalach'].mean(), linestyle='--');



# Add a title to the figure

fig.suptitle("Heart Disease Analysis", fontsize=20,
             fontweight='bold');
```
