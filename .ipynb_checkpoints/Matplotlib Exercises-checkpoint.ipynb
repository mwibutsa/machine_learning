```python
import matplotlib.pyplot as plt

# prepate data

x = [1,2,3,4,5]
y = np.random.randint(10,size=(5,1))

# set up figure and axes

fig, ax = plt.subplots(figsize=(10,6))


# add data to axes 

ax.plot(x,y)

ax.set(title="Random numbers", 
      xlabel='Random x numbers',
      ylabel='Random y numbers');
```

```python
food = {
    'Chicken': 1000,
    'Chips': 500,
    'Pizza': 800
}
```
