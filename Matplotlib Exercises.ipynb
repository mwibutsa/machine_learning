```python
car_sales
```

```python
car_sales['Total Sales'] = car_sales['Price'].cumsum()
```

```python
car_sales
```

```python
car_sales['Sales Date'] = pd.date_range('1/1/2020', 
                                        periods=len(car_sales))
```

```python
car_sales
```

```python
car_sales = car_sales.rename(columns={'Sales Date': 'Sale Date'})
```

```python
car_sales
```

```python
fig, ax = plt.subplots()

ax.plot(car_sales['Sale Date'], car_sales['Total Sales'])
```

```python
heart_disease = pd.read_csv('heart-disease.csv')
```

```python
heart_disease.head()
```

```python
heart_disease['age'].hist(bins=50);
```

```python
heart_disease.plot.hist(subplots=True)
```

```python
heart_disease.plot.bar()
```

```python
heart_disease.plot.hist(subplots=True, figsize=(10,30));
```

```python
over_50 = heart_disease[heart_disease['age'] > 50]
```

```python
over_50
```

```python
plt.style.use('seaborn-whitegrid')
fig, ax = plt.subplots()

scatter = ax.scatter(x=over_50['age'], 
                     y=over_50['chol'],
                     c=over_50['target'])

ax.set(title='Heart desease cholesterol levels',
      xlabel='Age',
      ylabel='Cholesterol')

ax.axhline(over_50['chol'].mean(), linestyle='--')
```

```python
fig, ax = plt.subplots()

scatter  = ax.scatter(x=over_50['age'], 
                     y=over_50['chol'],
                     c=over_50['target'],
                     cmap='winter')

ax.set(title='Heart disease and cholersterol Levels',
      xlabel='Age',
      ylabel='Cholesterol')

ax.axhline(over_50['chol'].mean(), linestyle='--')
ax.legend(*scatter.legend_elements(),title='Target');
```
